**Basado en el análisis de las variables (ventas, margen, unidades vendidas) y la matriz de correlación (si se llegase a calcular en un análisis posterior), ¿cuáles crees que son los principales factores que contribuyen a la ganancia operativa?**

_Esta pregunta es abierta: reflexiona sobre la relación entre precio por unidad, unidades vendidas, ventas totales y margen operativo, y plantea posibles estrategias de negocio basadas en estos insights._

Los factores que generan mayor ganancia operativa segun el análisis que obtuve de los datos, son los siguientes:

a. Unidades Vendidas: Nos damos cuenta que a mayor cantidad de unidades vendidas, mayor será la ganancia operativa, siempre que el precio supere los costos de producción, por esto crear estrategias como promociones o expansión de mercados pueden aumentar el volumen de ventas.
b. Precio por Unidad: Vimos que un precio más alto incrementa la ganancia por unidad, pero si es demasiado elevado, puede reducir la demanda, por esto es importante encontrar un equilibrio en los precios ofertados.
c. Ventas Totales: Ya que estas se obtienen de multiplicar las unidades vendidas por el precio por unidad, un incremento en cualquiera de estos factores aumentará las ventas. Esto generara más ingresos, pero su impacto en la ganancia depende del margen operativo, como lo vimos anteriormente.
d. Margen Operativo: Nos mostro el porcentaje de las ventas que se convierten en ganancia después de restar costos operativos, tambien nos dimos cuenta  que un margen alto indica eficiencia en costos y precios bien ajustados.

Al relacionar estos conceptos obtenemos ques el Precio por Unidad con las Unidades Vendidas nos muestra que un aumento en el precio puede reducir la cantidad vendida y un precio bajo puede aumentar el volumen de ventas pero reducir la rentabilidad por unidad. Tambien las Ventas Totales con el Margen Operativo nos dice que a mayores ventas mayores ingresos, pero si los costos aumentan proporcionalmente, el margen no aumentara. Finalmente el Precio por Unidad con el Margen Operativo nos dice que un precio más alto mejora el margen si los costos se mantienen estables.
